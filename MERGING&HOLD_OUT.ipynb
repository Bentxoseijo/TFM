```python
kc_counter = Counter()

# Iterar sobre la columna de KC y contar las ocurrencias de cada skill
for skills in df_base_datos['KC(Default)'].dropna():
    for skill in skills.split('~~'):
        kc_counter[skill] += 1

# Convertir el Counter a un DataFrame para una mejor visualización
kc_count_df = pd.DataFrame(kc_counter.items(), columns=['Skill', 'Count']).sort_values(by='Count', ascending=False)
```

## Creación del merged_df2 de test

```python
merged_df2['Problem_Step'] = merged_df2['Problem Name'] + '-' + merged_df2['Step Name']

# Contar la frecuencia de cada combinación 'Problem_Step'
problem_step_counts = merged_df2['Problem_Step'].value_counts()

# Filtrar combinaciones que se hayan realizado más de 10 veces
frequent_problem_steps = problem_step_counts[problem_step_counts > 10].index

# Filtrar el merged_df2Frame original para mantener solo las filas con combinaciones frecuentes
filtered_merged_df2 = merged_df2[merged_df2['Problem_Step'].isin(frequent_problem_steps)]

filtered_merged_df2 = filtered_merged_df2.drop(columns='Problem_Step')
merged_df2 = merged_df2.drop(columns='Problem_Step')

# Seleccionar aleatoriamente alrededor de 600 filas de las combinaciones filtradas
test_merged_df2 = filtered_merged_df2.sample(n=800, random_state=42)

# Eliminar las filas seleccionadas del merged_df2Frame original
merged_df2 = merged_df2.drop(test_merged_df2.index)

```

```python
# Paso 1: Dividir las celdas en listas usando el delimitador '~~'
merged_df2['KC(Default)'] = merged_df2['KC(Default)'].apply(lambda x: x.split('~~') if x != 'nan' else [])
merged_df2['Opportunity(Default)'] = merged_df2['Opportunity(Default)'].apply(lambda x: x.split('~~') if x != 'nan' else [])

# Número máximo de skills y oportunidades
max_kc_len = 5
max_opportunity_len = 5

# Paso 2: Expandir en columnas separadas y rellenar con ceros
for i in range(max_kc_len):
    merged_df2[f'KC_{i+1}'] = merged_df2['KC(Default)'].apply(lambda x: x[i] if i < len(x) else 0)

for i in range(max_opportunity_len):
    merged_df2[f'Opportunity_{i+1}'] = merged_df2['Opportunity(Default)'].apply(lambda x: x[i] if i < len(x) else 0)

# Eliminar columnas originales si no son necesarias
merged_df2.drop(columns=['KC(Default)', 'Opportunity(Default)'], inplace=True)


# Paso 1: Dividir las celdas en listas usando el delimitador '~~'
test_merged_df2['KC(Default)'] = test_merged_df2['KC(Default)'].apply(lambda x: x.split('~~') if x != 'nan' else [])
test_merged_df2['Opportunity(Default)'] = test_merged_df2['Opportunity(Default)'].apply(lambda x: x.split('~~') if x != 'nan' else [])

# Número máximo de skills y oportunidades
max_kc_len = 5
max_opportunity_len = 5

# Paso 2: Expandir en columnas separadas y rellenar con ceros
for i in range(max_kc_len):
    test_merged_df2[f'KC_{i+1}'] = test_merged_df2['KC(Default)'].apply(lambda x: x[i] if i < len(x) else 0)

for i in range(max_opportunity_len):
    test_merged_df2[f'Opportunity_{i+1}'] = test_merged_df2['Opportunity(Default)'].apply(lambda x: x[i] if i < len(x) else 0)

# Eliminar columnas originales si no son necesarias
test_merged_df2.drop(columns=['KC(Default)', 'Opportunity(Default)'], inplace=True)
```

```python
merged_df2 = merged_df2.applymap(lambda x: x.replace(' ', '_') if isinstance(x, str) else x)
test_merged_df2 = test_merged_df2.applymap(lambda x: x.replace(' ', '_') if isinstance(x, str) else x)
```
